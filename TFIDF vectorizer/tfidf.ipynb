```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OneHotEncoder,LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv("Salaries.csv")
df.head()
```

```python
df.info()
```

```python
df = df.drop(['Notes'],axis=1)
```

```python
df['Status'] = df['Status'].fillna('unknown')
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df[['BasePay','OvertimePay','OtherPay','Benefits']] = df[['BasePay','OvertimePay','OtherPay','Benefits']].replace('Not Provided','0')
```

```python
df['Agency'] = df['Agency'].replace('San Francisco','0')
```

```python
df[['BasePay','OvertimePay','OtherPay','Benefits','Agency']] = df[['BasePay','OvertimePay','OtherPay','Benefits','Agency']].astype(dtype=float)
```

```python
df['Agency']
```

```python
tfidf = TfidfVectorizer()
val = tfidf.fit_transform(df['JobTitle']).toarray()
tf_df = pd.DataFrame(val,columns=tfidf.get_feature_names_out())
tf_df
```

```python
# tfidf.vocabulary_
```

```python
scaler = StandardScaler()
```

```python
num_col = df.select_dtypes(include='number').columns
num_col

tf_df[num_col] = scaler.fit_transform(df[num_col])
```

```python
encode = LabelEncoder()
```

```python
tf_df['Status'] = encode.fit_transform(df['Status'])
```

```python
tf_df
```

## Transformer

```python
pre = ColumnTransformer(
    transformers=[
        ('tf-idf',TfidfVectorizer(),'JobTitle')             # use standard scaler here
        
    ]
)
#  'JobTitle'  here is 0-D amd is a filter condition 
pre.fit_transform(df).toarray()
```
