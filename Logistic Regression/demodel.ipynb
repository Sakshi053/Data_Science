```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler

# Modified, more realistic data
ages = np.array([20, 22, 25, 28, 35, 40, 50]).reshape(-1, 1)
incomes = np.array([18000, 22000, 50000, 80000, 150000, 300000, 500000]).reshape(-1, 1)

# Combine into one dataset
data = np.hstack((ages, incomes))

# Apply standard scaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Plot before scaling
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(data[:, 0], data[:, 1], color='red')
plt.title("Before Scaling")
plt.xlabel("Age")
plt.ylabel("Income")

# Plot after scaling
plt.subplot(1, 2, 2)
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], color='green')
plt.title("After Scaling")
plt.xlabel("Age (scaled)")
plt.ylabel("Income (scaled)")

plt.tight_layout()
plt.show()
```
