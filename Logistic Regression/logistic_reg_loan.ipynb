```python
obj_cols = df.select_dtypes(include=['object']).columns
obj_cols
```

```python
encode= OneHotEncoder()
tv=encode.fit_transform(df[obj_cols])
tv.toarray()
encode.get_feature_names_out(obj_cols)
```

```python

obj_df=pd.DataFrame(tv.toarray(), columns=encode.get_feature_names_out(obj_cols))
obj_df
```

one hot encoder can be used only for categorical data with one word 


 concat numerical col and categorical col

```python

df=pd.concat([df.drop(obj_cols, axis=1), obj_df], axis=1)
df
```

```python
x= df.drop('loan_status', axis=1)
y= df['loan_status']

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)   
```

```python
model= LogisticRegression(max_iter=13000)
model.fit(xtrain, ytrain)
```

```python
model.n_iter_
```

```python
model.score(xtest, ytest)
```
