```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
height = np.array([150,152,155,158,160,162,165,168,170,173,175,178,180,182,185,188,189,190]).reshape(-1,1)
height
```

```python
model = KMeans(n_clusters=3,random_state=42)
model.fit(height)
```

```python
label = model.labels_
label
```

```python
centroid = model.cluster_centers_
centroid
```

```python
for i in range(len(height)):
    print(f"Height{height[i]}--->cm Group:{label[i]}")
```

```python
plt.scatter(height,label)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(height,np.zeros_like(height),c=label,s=100,label="Data Points")
plt.scatter(centroid,np.zeros_like(centroid),c='red',marker='X',s=200,label="Centroids")
plt.title('K-means Clustering of student heights')
plt.legend()
plt.grid()
plt.xlabel('Height (cm)')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(height,(height),c=label,s=100,label="Data Points")
plt.scatter(centroid,(centroid),c='red',marker='X',s=200,label="Centroids")
plt.title('K-means Clustering of student heights')
plt.legend()
plt.grid()
plt.xlabel('Height (cm)')
plt.show()
```

```python
wcss = []
for i in range(1,6):
    kmeans = KMeans(n_clusters=i,random_state=42)
    kmeans.fit(height)
    wcss.append(kmeans.inertia_)
wcss
```

```python
plt.plot(range(1,6),wcss,marker='o')
plt.grid(True)
plt.show()
```
